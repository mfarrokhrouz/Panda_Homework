```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# This code is used to get the brief information of data set
school_data_complete.info()
```

```python
# This code is used to determine the name of the columns
school_data_complete.columns
```

```python
# This section counts the total number of the unique school names
count = school_data_complete.school_name.unique()
print(f"The total number of schools is: ", len(count))
```

```python
# This section counts the total number of students
number_of_students = school_data_complete.student_name
print(f"The total number of students is: ", len(number_of_students))
```

```python
selection = school_data_complete.budget.unique()
budget = selection.sum()
print(f"The total schools budget is: $", budget)
```

```python
math = school_data_complete.maths_score.mean()
print(f"The average math score is: ", round(math, 2))
```

```python
reading = school_data_complete.reading_score.mean()
print(f"The average reading score is: ", round(reading, 2))
```

```python
# This section is for percentage of those who pass math
math_passing = school_data_complete.maths_score.loc[school_data_complete.maths_score > 49]
print(f"The percentage of students who passed math is", round((100*len(math_passing) / len(number_of_students)), 6), "%")
```

```python
# This section is for percentage of those who pass reading
reading_passing = school_data_complete.reading_score.loc[school_data_complete.reading_score > 49]
print(f"The percentage of students who passed reading is:", round((100*len(reading_passing) / len(number_of_students)), 6), "%")
```

```python
new_math = school_data_complete.maths_score
new_reading = school_data_complete.reading_score
both = 0
for i in range(0, len(number_of_students)):
    if new_math[i] > 49 and new_reading[i] > 49:
        both = both + 1
print(f"The percentage of students who passed both math and reading is:", round((100*both / len(number_of_students)), 6), "%")
```

```python
# Create a DataFrame of frames using a dictionary of lists
frame_df = pd.DataFrame({
    "Total Schools": [len(count)],
    "Total Students": [len(number_of_students)],
    "Total Budget ($)": [budget],
    "Average Math Mark": [math],
    "Average Mark Reading": [reading],
    "Percentage Passed Math (%)": [round((100*len(math_passing) / len(number_of_students)), 6)],
    "Percentage Reading Passed (%)": [round((100*len(reading_passing) / len(number_of_students)), 6)],
    "Percentage Passed Both Math and reading (%)": [round((100*both / len(number_of_students)), 6)]
})
frame_df
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
year_9_avg = school_data_complete.loc[school_data_complete.year == 9].groupby("school_name")["maths_score"].mean()
year_9_avg 
year_10_avg = school_data_complete.loc[school_data_complete.year == 10].groupby("school_name")["maths_score"].mean()
year_10_avg
year_11_avg = school_data_complete.loc[school_data_complete.year == 11].groupby("school_name")["maths_score"].mean()
year_11_avg
year_12_avg = school_data_complete.loc[school_data_complete.year == 12].groupby("school_name")["maths_score"].mean()
year_12_avg
years_avg_1 = pd.merge(year_9_avg, year_10_avg, on="school_name")
years_avg_1 = years_avg_1.rename(
    columns={"maths_score_x": "year 9"})
years_avg_1 = years_avg_1.rename(
    columns={"maths_score_y": "year 10"})
years_avg_1
years_avg_2 = pd.merge(years_avg_1, year_11_avg, on="school_name")
years_avg_3 = pd.merge(years_avg_2, year_12_avg, on="school_name")
years_avg_3 = years_avg_3.rename(columns={"maths_score_x": "year 11"})
years_avg_3 = years_avg_3.rename(columns={"maths_score_y":"year 12"})
years_avg_3
```
